```python
import numpy as np
import matplotlib.pyplot as plt

#simple plot
x = np.linspace(0,10,100)

fig = plt.figure()

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
fig.savefig('basicplot.png')
plt.show()
```

```python
#object oriented
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x), '--g');
```

```python
plt.plot(x, np.sin(x),'g-',label = 'sin(x)')
plt.title("A Sin Curve")
plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.axis([-1, 11, -1.5, 1.5])
plt.legend();
```
